```python
def goldbach1(m):
    pList = []
    g = range(4,m+1,2)
    f = [0]*len(g)
    for i in range(2, m):
        allnumbersless = range(2, i)
        prime = True
        for a in allnumbersless:
            n = i % a 
            if n == 0:
                prime = False
                break
        if prime:
            pList.insert(0,i)
    for j in range(len(g)):
        k = g[j]
        for b in pList:
            c = k-b
            if c > 1 and c in pList and c <= b:
                #print(k, "=", b, "+", c)
                f[j] += 1
    return f  
```

```python
m = 5000 
g = range(4,m+1,2)

import time
start_time = time.clock()

y1 = goldbach1(m)
print(round(time.clock() - start_time, 2), "seconds")
```

```python
from matplotlib import pyplot as plt
import matplotlib

s = 5

plt.scatter(g, y1, s, c="turquoise", alpha=0.5)
plt.xlabel("Even Numbers")
plt.ylabel("# of Pairs of Primes")
plt.legend(loc=2)
plt.show()
```
