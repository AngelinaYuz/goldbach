```python
m = 1000 
pList = []
for i in range(2, m):
    allnumbersless = range(2, i)
    prime = True
    for a in allnumbersless:
        n = i % a 
        if n == 0:
            prime = False
            break
    if prime:
        pList.insert(0,i)
for x in range(4,m+1,2):
    for b in pList:
        c = x-b
        if c > 1 and c in pList and c <= b:
            print(x, "=", b, "+", c)

#print(pList)
```
