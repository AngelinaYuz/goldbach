```python
def goldbach1(m):
    pList = []
    g = range(4,m+1,2)
    f = [0]*len(g)
    for i in range(2, m):
        allnumbersless = range(2, i)
        prime = True
        for a in allnumbersless:
            n = i % a 
            if n == 0:
                prime = False
                break
        if prime:
            pList.insert(0,i)
    for j in range(len(g)):
        k = g[j]
        for b in pList:
            c = k-b
            if c > 1 and c in pList and c <= b:
                #print(k, "=", b, "+", c)
                f[j] += 1
    return f  
```

```python
def goldbach2(m):
    pList = []
    g = range(4,m+1,2)
    f = [0]*len(g)
    for i in range(2, m):
        allnumbersless = range(2, int(i**0.5))
        prime = True
        for a in allnumbersless:
            n = i % a 
            if n == 0:
                prime = False
                break
        if prime:
            pList.append(i)
    for j in range(len(g)):
        k = g[j]
        i = 0
        b = 2
        while (b < k - 1) and (i < len(pList)):
            b = pList[i]
            c = k-b
            if c in pList and c <= b:
                #print(k, "=", b, "+", c)
                f[j] += 1
            i = i+1
    return f  
```

```python
m = 4000 
g = range(4,m+1,2)
import time
qList = []
kList = []

for q in range(200, m+1, 200):
    start_time = time.clock()
    y1 = goldbach2(q)
    k = round(time.clock() - start_time, 2)
    qList.insert(0,q)
    kList.insert(0,k)
    #print(q, ":", k, "seconds")
```

```python
from matplotlib import pyplot as plt
import matplotlib

h = 10

plt.scatter(qList, kList, h, c="red", alpha=0.5)
plt.xlabel("Measurement (Even Number)")
plt.ylabel("Processing Time")
plt.legend(loc=2)
plt.show()
```
